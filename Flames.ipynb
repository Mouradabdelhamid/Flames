```python
def flames(n1, n2):
  for i in n1:
    for j in n2:
      n1 = n1.replace(i, "", 1)
      n2 = n2.replace(j, "", 1)
  n3 = len(n1 + n2)
  return n3
x = input('Please enter your name : ')
y = input("Please enter the second person's name : ")
z = flames(x, y)
if z >= 0:
  flamelist = ['Friends', 'Lovers', 'Affectionate', 'Marriage', 'Enemies', 'Siblings']
  while len(flamelist) > 1:
    count = z % len(flamelist)
    split = count - 1
    if split >= 0:
      right = flamelist[split + 1:]
      left = flamelist[:split]
      flamelist = right + left
    else:
      flamelist = flamelist[:len(flamelist) - 1]
  print('Your relationship with the second person is ',flamelist[0])
else:
  print('Enter other names')
```
